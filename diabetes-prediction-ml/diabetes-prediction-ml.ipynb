```python
#Train and Test split
from sklearn.model_selection import train_test_split
```

```python
x=df.loc[:,df.columns!='Outcome'] # features (everything except 'Outcome')
#y=df.loc[:,df.columns=='Outcome']  target (only 'Outcome') ❌ This returns a **DataFrame**, not a Series
y = df['Outcome']
```

```python
print(x) #prints data without outcome
```

```python
print(y) #prints only outcome
```

```python
# Explanation of Variables after Train-Test Split:
# xtrain → 80% of the feature data used for training the model
# xtest  → 20% of the feature data used for testing the model
# ytrain → 80% of the target labels used to train the model
# ytest  → 20% of the target labels used to evaluate the model
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2) # Train-test split
```

```python
# Importing RandomForestClassifier from scikit-learn to build a classification model
from sklearn.ensemble import RandomForestClassifier 
```

```python
model=RandomForestClassifier()
```

```python
model.fit(xtrain,ytrain)
```

```python
outputprediction=model.predict(xtest)
print(outputprediction)
```

```python
from sklearn.metrics import accuracy_score
```

```python
score=accuracy_score(outputprediction,ytest)
print(score)
```
